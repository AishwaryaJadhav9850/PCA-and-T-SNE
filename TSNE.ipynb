```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE


data1=pd.read_csv("train.csv")
l=data1['label']
d=data1.drop("label",axis=1)
labels=l.head(25000)
data=d.head(25000)
print(labels.shape)
print(data.shape)
```

# Column Standardization

```python
std_data = StandardScaler().fit_transform(data)
print(std_data.shape)
```

```python
model=TSNE(n_components=2,random_state=0,perplexity=30,n_iter=1000)
tsne_data=model.fit_transform(std_data)
```

```python

tsne_data=np.vstack((tsne_data.T,labels)).T

df=pd.DataFrame(data=tsne_data,columns=("Dim_1", "Dim_2", "label_name"))

sns.FacetGrid(df, hue="label_name", size=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```
